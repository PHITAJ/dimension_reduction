    2. Calculer la distance euclidienne des 5 plus proches voisins au point d'indice zéro. Est-ce en accord avec l'observation visuelle précédente ?


Plusieurs fonctions permettent de calculer facilement ces distances. Dans le module linalg de numpy, on peut utiliser la fonction <a ref="https://numpy.org/doc/stable/reference/generated/numpy.linalg.norm.html">norm</a>
</h5>

```python
point_zero_X1 = X1[0]
```

```python
distances = []
for i in X1:
   distances.append(np.linalg.norm(i - point_zero_X1))
```

```python
distances.sort()
```

```python
distances[1:6]
```

```python
X2d = np.concatenate((X1.reshape(100,1), X2.reshape(100,1)), axis=1)
```

```python
point_zero_X2d = X2d[0]
```

```python
distances_2d = []
for i in X2d:
   distances_2d.append(np.linalg.norm(i - point_zero_X2d))
```

```python
distances_2d.sort()
```

```python
distances_2d[1:6]
```

```python
X3d = np.concatenate((X1.reshape(100,1), X2.reshape(100,1),X3.reshape(100,1)), axis=1)
```

```python
point_zero_X3d = X3d[0]
```

```python
distances_3d = []
for i in X3d:
   distances_3d.append(np.linalg.norm(i - point_zero_X3d))
```

```python
distances_3d.sort()
```

```python
distances_3d[1:6]
```
