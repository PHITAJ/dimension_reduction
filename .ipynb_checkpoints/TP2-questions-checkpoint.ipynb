# PCA from scratch

## En s'aidant de sklearn

````python
# some utils
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import eigh
from sklearn.decomposition import PCA
from sklearn import datasets
````

````python
mean = [2., 3.]
cov = [[.01,0],
        [0, .1]
        ]
n = 1000
M1 = [[1,1], [-1, 1]]
M2 = [[0.6, -0.6], [-0.4, 0.8]]
points = np.random.multivariate_normal(mean, cov, n)

np.random.seed(1)

blob1 = (M1 @ points.T).T
blob2 = (M2 @ points.T).T
blob3, yblob3 = datasets.make_blobs(n_samples=n)
cercles, ycercles = datasets.make_circles(n_samples=n, factor=0.5, noise=0.05)
lunes, ylunes = datasets.make_moons(n_samples=n, noise=0.05)
````

## Implémentation maison

Le but de cette partie est de recoder la PCA et de vérifier les résultat grâce à la comparaison avec le module sklearn. Il faut faire chaque question pour chaque jeu de données.

<h3 style="background-color: #ffa0b6">
   3. Centrer les données
</h3>

````python
blob1_centered = blob1 - np.mean(blob1, axis=0)
````

````python
blob1
````

````python
blob1_centered
````

````python
blob2_centered = blob2 - np.mean(blob2, axis=0)
````

````python
blob3_centered = blob3 - np.mean(blob3, axis=0)
````

````python
cercles_centered = cercles - np.mean(cercles, axis=0)
````

````python
lunes_centered = lunes - np.mean(lunes, axis=0)
````

<h3 style="background-color: #ffa0b6">
   4. Calculer la matrice de covariance empirique $\Sigma$
</h3>

````python
cov_blob1 = (1/n) * (blob1_centered.T @ blob1_centered)
````

````python
cov_blob1
````

````python
cov_blob2 = (1/n) * (blob2_centered.T @ blob2_centered)
````

````python
cov_blob3 = (1/n) * (blob3_centered.T @ blob3_centered)
````

````python
cov_cercles = (1/n) * (cercles_centered.T @ cercles_centered)
````

````python
cov_lunes = (1/n) * (lunes_centered.T @ lunes_centered)
````

<h3 style="background-color: #ffa0b6">
   5. Calculer les valeurs propres et les vecteurs propres
</h3>

````python
valeurs_propres_blob1, vecteurs_propres_blob1 = eigh(cov_blob1)
````

````python
valeurs_propres_blob2, vecteurs_propres_blob2 = eigh(cov_blob2)
````

````python
valeurs_propres_blob3, vecteurs_propres_blob3 = eigh(cov_blob3)
````

````python
valeurs_propres_cercles, vecteurs_propres_cercles = eigh(cov_cercles)
````

````python
valeurs_propres_lunes, vecteurs_propres_lunes = eigh(cov_lunes)
````

<h3 style="background-color: #ffa0b6">
   6. Exprimer $\Sigma$ en fonction des valeurs et vecteurs propres précedemment calculés et vérifier empiriquement l'égalité
</h3>

````python
vecteurs_propres_blob1 @ (valeurs_propres_blob1 * np.eye(2)) @ vecteurs_propres_blob1.T
````

````python
cov_blob1
````

````python
vecteurs_propres_blob2 @ (valeurs_propres_blob2 * np.eye(2)) @ vecteurs_propres_blob2.T
````

````python
cov_blob2
````

````python
vecteurs_propres_blob3 @ (valeurs_propres_blob3 * np.eye(2)) @ vecteurs_propres_blob3.T
````

````python
cov_blob3
````

````python
vecteurs_propres_cercles @ (valeurs_propres_cercles * np.eye(2)) @ vecteurs_propres_cercles.T
````

````python
cov_cercles
````

````python
vecteurs_propres_lunes @ (valeurs_propres_lunes * np.eye(2)) @ vecteurs_propres_lunes.T
````

````python
cov_lunes
````
