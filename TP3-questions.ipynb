## PCA sur un jeu de données de plus grandes dimensions

   1. Télécharger MNIST et afficher un chiffre du jeu de données.

En cas de difficulté (ordinateur trop lent) on pourra se contenter d'utiliser la version simplifiée de MNIST ``digits`` présente dans sklearn. MNIST peut être téléchargé facilement via ``fetch_openml``

```python
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import scale, minmax_scale
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import torch
import pandas as pd
```

```python
X, y = fetch_openml('mnist_784', version=1, return_X_y=True, as_frame=False)
A = pd.DataFrame(X)
A['y'] = y
df = A.drop("y", axis=1)
```

```python
X = torch.tensor(minmax_scale(A.drop("y", axis=1)).astype(np.float32)).reshape(A.shape[0],1,28,28)
```

```python
plt.imshow(X[0][0], cmap="Greys")
plt.colorbar()
```

```python
y[0][0]
```

   2. Appliquer PCA en dimension 2 et afficher le résultat et colorant les points selon le chiffre qu'ils représentent.

On utilisera la version de ``sklearn``.

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
df_pca = pca.fit_transform(df)
```

```python
df_pca[:,0]
```

```python
plt.scatter(df_pca[:,0],df_pca[:,1], c=y.astype(int), cmap="tab10")
```

   3. Afficher le barycentre des points de chaque chiffre dans la projection en deux dimensions.

On attend donc une figure avec la projection du jeu de données en deux dimensions, et en plus 10 points/marqueurs clairement identifiables correspondant chacun à un chiffre, de la couleur du chiffre qu'il représente.

```python
barycentres = {}
barycentres = {label: [] for label in range(10)}
for valeurs, label in zip(df_pca, y):
    barycentres[int(label)].append(valeurs)
```

```python
for i in barycentres :
    barycentres[i] = np.mean(barycentres[i], axis=0)
```

```python
plt.scatter(np.array(list(barycentres.values()))[:,0], np.array(list(barycentres.values()))[:,1], c=range(10), cmap="tab10")
```

   4. Afficher les barycentres précédemment calculés en les retransformant dans l'espace initial, c'est-à-dire en affichant les images de chacun des 10 chiffres.

```python
transfo = pca.inverse_transform(np.array(list(barycentres.values())))
```

```python
transfo_reschape = []
for i in transfo:
    transfo_reschape.append(i.reshape(28,28))
```

```python
fig, axes = plt.subplots(1,10, figsize=(40,16))
for i, ax in enumerate(axes.flat):
    ax.imshow(transfo_reschape[i], cmap="Greys")
plt.show()
```
